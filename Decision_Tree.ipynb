```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
dataset=pd.read_csv('/content/Social_Network_Ads.csv')
```

```python
dataset
```

```python
dataset.corr()
```

```python
dataset.drop(['User ID','Gender'],inplace=True,axis=1)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler=StandardScaler()
```

```python
scaler.fit(dataset.drop('Purchased',axis=1))
```

```python
scaled_features=scaler.transform(dataset.drop('Purchased',axis=1))
```

```python
scaled_features
```

```python
df=pd.DataFrame(scaled_features,columns=['Age','EstimatedSalary'])
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test= train_test_split(df,dataset['Purchased'],test_size=0.30,random_state=60)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
DTC=DecisionTreeClassifier()
```

```python
DTC.fit(x_train,y_train)
```

```python
pred=DTC.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
print(confusion_matrix(y_test,pred))
```
